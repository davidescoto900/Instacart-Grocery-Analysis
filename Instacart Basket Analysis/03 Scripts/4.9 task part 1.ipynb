```python
def label_price_range(price):
    if price < 5:
        return 'Low-range product'
    elif price <= 15:
        return 'Mid-range product'
    else:
        return 'High-range product'

full_merged_df['price_range_loc'] = full_merged_df['prices'].apply(label_price_range)
```

```python
full_merged_df.isnull().sum()
```

```python
# In the original 'price_range_loc' column, some products with a price of exactly $5.00
# were not labeled, resulting in missing values (NaNs). After reviewing the price distribution
# for each label, it became clear that the $5.00 value wasn’t included in any of the defined ranges.
#
# To fix this, I redefined the price labeling logic:
# - Products under $5.00 are labeled 'Low-range product'
# - Products from $5.00 to $15.00 are labeled 'Mid-range product'
# - Products above $15.00 are labeled 'High-range product'
```
